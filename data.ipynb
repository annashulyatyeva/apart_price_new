```python
#СОЗДАДИМ СПИСКИ ДЛЯ СОРТИРОВКИ ПО РАЙОНУ (ГОРОДА ИЛИ ОБЛАСТИ)
district_list = list(data.district.unique())
district_list
```

```python
district_list = list(data.district.unique())
#district_list
obl_list = []
town_list = []
for i in district_list:  
    if len(data[data.district == i]) <= 6:
        obl_list.append(i)
    elif len(data[data.district == i]) < 20 and len(data[data.district == i]) > 6:        
        town_list.append(i)
    else: pass    
nn_district_list = list(set(district_list) - set(obl_list) - set(town_list))
print(obl_list)
print("\n")
print(town_list)
print("\n")
print(nn_district_list)
```

```python
year_list = list(data.year.unique())
#year_list
```

```python
second_hand_list = []
new_build_list = []
for i in year_list:  
    if i < 2024.0:
        second_hand_list.append(i)
    else:       
        new_build_list.append(i)
print(second_hand_list)
print("\n")
print(new_build_list)
```
